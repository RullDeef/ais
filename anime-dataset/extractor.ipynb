```python
import pandas as pd

# load AnimeList.csv
df = pd.read_csv('anime_cleaned.csv')
df
```

```python
genres = dict()
for id, vals in df[['anime_id', 'genre']].values:
    if type(vals) is not str:
        continue
    for gnr in vals.split(', '):
        if gnr not in genres.keys():
            genres[gnr] = [id]
        else:
            genres[gnr].append(id)

for key, val in genres.items():
    print(key, val)
```

```python
reviews_df = pd.read_csv('users_filtered.csv')

reviews_df
```

```python
# extract relations graph

import json

def fix_quotes(s: str) -> str:
    s = s.replace('"', '__QUOTE__')
    s = s.replace("'", '"')
    s = s.replace('__QUOTE__', "'")
    return s

graph = dict()

for id in df['anime_id']:
    graph[id] = {'edges': [], 'edges_count': 0}

for id, related in df[['anime_id', 'related']].values:
    if related == '[]':
        continue
    rel = json.loads(fix_quotes(related))
    for typ, vals in rel.items():
        for item in vals:
            graph[id]['edges'].append({
                'id': item['mal_id'],
                'type': item['type'],
                'relation': typ,
            })
            graph[id]['edges_count'] += 1

graph
```
