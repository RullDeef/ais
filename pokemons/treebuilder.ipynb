```python
import pandas as pd
import numpy as np

# use only given columns
dataset = pd.read_csv('pokemon.csv')
dataset = dataset[['name', 'classfication', 'type1', 'type2', 'hp', 'attack', 'speed', 'abilities']]

classes = dataset['classfication'].unique()

print('all dataset size:', len(dataset))
print('amount of different classes:', len(classes))

classes_distr = [(c, len(dataset.loc[dataset['classfication'] == c])) for c in classes]

# reduce dataset to pokemons which form at least class of 4 pokies
classes_distr = list(filter(lambda s: s[1] >= 4, classes_distr))
print('filtered classes:', len(classes_distr))

classes_distr = sorted(classes_distr, key=lambda s: -s[1])
classes_distr

classes = [s[0] for s in classes_distr]
classes
```

```python
# determine spread for each class

for clazz in classes:
    X = dataset.loc[dataset['classfication'] == clazz][['type1', 'type2', 'hp', 'attack', 'speed', 'abilities']]
    X = pd.get_dummies(X)
    
```
